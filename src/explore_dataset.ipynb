```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/rent_commute_data.csv")
df.head()
```

```python
df.describe()
```

## Rent vs Life_Cost

```python
plt.figure(figsize=(6,4))
plt.scatter(df['rent'], df['life_cost'], color='blue')
plt.title("Rent vs Life Cost")
plt.xlabel("Rent (₹)")
plt.ylabel("Life Cost (₹)")
plt.show()
```

# Commute vs Life_Cost

```python
plt.figure(figsize=(6,4))
plt.scatter(df['commute_minutes'], df['life_cost'], color='green')
plt.title("Commute Minutes vs Life Cost")
plt.xlabel("Commute (minutes/day)")
plt.ylabel("Life Cost (₹)")
plt.show()
```

# Correlation

```python
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
```
